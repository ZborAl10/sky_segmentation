```
from collections import OrderedDict

import torch
import torch.nn as nn


class UNet(nn.Module):

    def __init__(self, in_channels=3, out_channels=1, init_features=32):
        super(UNet, self).__init__()

        features = init_features
        self.encoder1 = UNet._block(in_channels, features, name="enc1")
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.encoder2 = UNet._block(features, features * 2, name="enc2")
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.encoder3 = UNet._block(features * 2, features * 4, name="enc3")
        self.pool3 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.encoder4 = UNet._block(features * 4, features * 8, name="enc4")
        self.pool4 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.bottleneck = UNet._block(features * 8, features * 16, name="bottleneck")

        self.upconv4 = nn.ConvTranspose2d(
            features * 16, features * 8, kernel_size=2, stride=2
        )
        self.decoder4 = UNet._block((features * 8) * 2, features * 8, name="dec4")
        self.upconv3 = nn.ConvTranspose2d(
            features * 8, features * 4, kernel_size=2, stride=2
        )
        self.decoder3 = UNet._block((features * 4) * 2, features * 4, name="dec3")
        self.upconv2 = nn.ConvTranspose2d(
            features * 4, features * 2, kernel_size=2, stride=2
        )
        self.decoder2 = UNet._block((features * 2) * 2, features * 2, name="dec2")
        self.upconv1 = nn.ConvTranspose2d(
            features * 2, features, kernel_size=2, stride=2
        )
        self.decoder1 = UNet._block(features * 2, features, name="dec1")

        self.conv = nn.Conv2d(
            in_channels=features, out_channels=out_channels, kernel_size=1
        )

    def forward(self, x):
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool1(enc1))
        enc3 = self.encoder3(self.pool2(enc2))
        enc4 = self.encoder4(self.pool3(enc3))

        bottleneck = self.bottleneck(self.pool4(enc4))

        dec4 = self.upconv4(bottleneck)
        dec4 = torch.cat((dec4, enc4), dim=1)
        dec4 = self.decoder4(dec4)
        dec3 = self.upconv3(dec4)
        dec3 = torch.cat((dec3, enc3), dim=1)
        dec3 = self.decoder3(dec3)
        dec2 = self.upconv2(dec3)
        dec2 = torch.cat((dec2, enc2), dim=1)
        dec2 = self.decoder2(dec2)
        dec1 = self.upconv1(dec2)
        dec1 = torch.cat((dec1, enc1), dim=1)
        dec1 = self.decoder1(dec1)
        return torch.sigmoid(self.conv(dec1))

    @staticmethod
    def _block(in_channels, features, name):
        return nn.Sequential(
            OrderedDict(
                [
                    (
                        name + "conv1",
                        nn.Conv2d(
                            in_channels=in_channels,
                            out_channels=features,
                            kernel_size=3,
                            padding=1,
                            bias=False,
                        ),
                    ),
                    (name + "norm1", nn.BatchNorm2d(num_features=features)),
                    (name + "relu1", nn.ReLU(inplace=True)),
                    (
                        name + "conv2",
                        nn.Conv2d(
                            in_channels=features,
                            out_channels=features,
                            kernel_size=3,
                            padding=1,
                            bias=False,
                        ),
                    ),
                    (name + "norm2", nn.BatchNorm2d(num_features=features)),
                    (name + "relu2", nn.ReLU(inplace=True)),
                ]
            )
        )



#UNet architecture code is takn from https://github.com/mateuszbuda/brain-segmentation-pytorch
```

```
class Discriminator(nn.Module):
    def __init__(self, in_channels, use_sigmoid=True, use_spectral_norm=True):
        super().__init__()
        self.use_sigmoid = use_sigmoid

        self.conv1 = self.features = nn.Sequential(
            spectral_norm(nn.Conv2d(in_channels=in_channels, out_channels=64, kernel_size=4, stride=2, padding=1, bias=not use_spectral_norm), use_spectral_norm),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv2 = nn.Sequential(
            spectral_norm(nn.Conv2d(in_channels=64, out_channels=128, kernel_size=4, stride=2, padding=1, bias=not use_spectral_norm), use_spectral_norm),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv3 = nn.Sequential(
            spectral_norm(nn.Conv2d(in_channels=128, out_channels=256, kernel_size=4, stride=2, padding=1, bias=not use_spectral_norm), use_spectral_norm),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv4 = nn.Sequential(
            spectral_norm(nn.Conv2d(in_channels=256, out_channels=512, kernel_size=4, stride=1, padding=1, bias=not use_spectral_norm), use_spectral_norm),
            nn.LeakyReLU(0.2, inplace=True),
        )

        self.conv5 = nn.Sequential(
            spectral_norm(nn.Conv2d(in_channels=512, out_channels=1, kernel_size=4, stride=1, padding=1, bias=not use_spectral_norm), use_spectral_norm),
        )


    def forward(self, x):
        conv1 = self.conv1(x)
        conv2 = self.conv2(conv1)
        conv3 = self.conv3(conv2)
        conv4 = self.conv4(conv3)
        conv5 = self.conv5(conv4)

        outputs = conv5
        if self.use_sigmoid:
            outputs = torch.sigmoid(conv5)

        return outputs, [conv1, conv2, conv3, conv4, conv5]



def spectral_norm(module, mode=True):
    if mode:
        return nn.utils.spectral_norm(module)

    return module


#discriminator code is taken from https://github.com/knazeri/edge-informed-sisr
```

```
class AdversarialLoss(nn.Module):
    r"""
    Adversarial loss
    https://arxiv.org/abs/1711.10337
    """

    def __init__(self, type='hinge', target_real_label=1.0, target_fake_label=0.0):
        r"""
        type = nsgan | lsgan | hinge
        """
        super().__init__()

        self.type = type
        self.register_buffer('real_label', torch.tensor(target_real_label))
        self.register_buffer('fake_label', torch.tensor(target_fake_label))

        if type == 'nsgan':
            self.criterion = nn.BCELoss()

        elif type == 'lsgan':
            self.criterion = nn.MSELoss()

        elif type == 'hinge':
            self.criterion = nn.ReLU()

    def __call__(self, outputs, is_real, is_disc=None):
        if self.type == 'hinge':
            if is_disc:
                if is_real:
                    outputs = -outputs
                return self.criterion(1 + outputs).mean()
            else:
                return (-outputs).mean()

        else:
            labels = (self.real_label if is_real else self.fake_label).expand_as(outputs)
            loss = self.criterion(outputs, labels)
            return loss

#Adversarial Loss code is taken from https://github.com/knazeri/edge-informed-sisr
```

```
class GU_Net(nn.Module):

    def __init__(self, unet_input, unet_output, discriminator_input, unet_dil_input, unet_dil_output, discriminator_dil_input):
        super().__init__()


        unet = UNet(in_channels=unet_input ,out_channels=unet_output)
        discriminator = Discriminator(in_channels=discriminator_input , use_sigmoid=True)
        unet_dil = UNet(in_channels=unet_input ,out_channels=unet_output)
        discriminator_dil = Discriminator(in_channels=discriminator_input , use_sigmoid=True)

        criterion = smp.losses.DiceLoss(mode='binary')
        adversarial_loss = AdversarialLoss(type='hinge')

        self.add_module('unet', unet)
        self.add_module('discriminator', discriminator)
        self.add_module('unet_dil', unet)
        self.add_module('discriminator_dil', discriminator)

        self.add_module('criterion', criterion)
        self.add_module('adversarial_loss', adversarial_loss)

        self.unet_optimizer = optim.Adam(
            unet.parameters(),
            lr = float(0.01)
            )

        self.dis_optimizer = optim.Adam(
             params=discriminator.parameters(),
             lr=float(0.00001),
             betas=(0.9, 0.999)
             )

        self.unet_dil_optimizer = optim.Adam(
            unet_dil.parameters(),
            lr = float(0.01)
            )

        self.dis_dil_optimizer = optim.Adam(
             params=discriminator.parameters(),
             lr=float(0.00001),
             betas=(0.9, 0.999)
             )

        self.unet_input = unet_input
        self.unet_output = unet_output
        self.discriminator_input = discriminator_input

        self.unet_dil_input = unet_dil_input
        self.unet_dil_output = unet_dil_output
        self.discriminator_dil_input = discriminator_dil_input


    def load(self, path_unet, path_discriminator, path_unet_dil, path_discriminator_dil):
        weight_unet = torch.load(path_unet)
        weight_discriminator = torch.load(path_discriminator)
        weight_unet_dil = torch.load(path_unet)
        weight_discriminator_dil = torch.load(path_discriminator)

        self.unet.load_state_dict(weight_unet)
        self.discriminator.load_state_dict(weight_discriminator)
        self.unet_dil.load_state_dict(weight_unet_dil)
        self.discriminator_dil.load_state_dict(weight_discriminator_dil)

    def save_weight(self, path_unet, path_dis, path_unet_dil, path_dis_dil):
        torch.save(self.unet.state_dict(), path_unet)
        torch.save(self.discriminator.state_dict(), path_dis)
        torch.save(self.unet_dil.state_dict(), path_unet_dil)
        torch.save(self.discriminator_dil.state_dict(), path_dis_dil)

    def process(self, rgb_images, mask_images, edge_images):

        # zero optimizers
        self.unet_optimizer.zero_grad()
        self.dis_optimizer.zero_grad()


        # find output and initialize loss to zero
        unet_loss = 0
        dis_loss = 0

        input_unet = torch.cat((rgb_images.cuda(), edge_images.cuda()), dim=1).cuda()
        outputs = self.unet(input_unet)


        # discriminator loss
        dis_input_real = torch.cat((rgb_images.cuda(), mask_images.cuda()), dim=1).cuda()
        dis_input_fake = torch.cat((rgb_images.cuda(), outputs.detach().cuda()), dim=1).cuda()
        dis_real, dis_real_feat = self.discriminator(dis_input_real)
        dis_fake, dis_fake_feat = self.discriminator(dis_input_fake)
        dis_real_loss = self.adversarial_loss(dis_real, True, True)
        dis_fake_loss = self.adversarial_loss(dis_fake, False, True)
        dis_loss += (dis_real_loss + dis_fake_loss) / 2


        # unet loss
        unet_input_fake = torch.cat((rgb_images.cuda(), outputs.cuda()), dim=1).cuda()
        unet_fake, unet_fake_feat = self.discriminator(unet_input_fake)
        unet_gan_loss = self.adversarial_loss(unet_fake, True, False) * 1.0
        #print('BBBB ' + str(unet_gan_loss.cpu().item()))
        unet_loss += unet_gan_loss

        unet_criterion = self.criterion(outputs.cuda(), mask_images.cuda())
        unet_loss += unet_criterion

        return unet_loss, dis_loss

    def process_dil(self, rgb_images, mask_images, pred_mask_images):

        # zero optimizers
        self.unet_dil_optimizer.zero_grad()
        self.dis_dil_optimizer.zero_grad()


        # find output and initialize loss to zero
        unet_loss_dil = 0
        dis_loss_dil = 0

        input_unet_dil = torch.cat((rgb_images.cuda(), pred_mask_images.cuda()), dim=1).cuda()
        outputs_dil = self.unet_dil(input_unet_dil)


        # discriminator loss
        dis_input_real = torch.cat((rgb_images.cuda(), mask_images.cuda()), dim=1).cuda()
        dis_input_fake = torch.cat((rgb_images.cuda(), outputs_dil.detach().cuda()), dim=1).cuda()
        dis_real, dis_real_feat = self.discriminator_dil(dis_input_real)
        dis_fake, dis_fake_feat = self.discriminator_dil(dis_input_fake)
        dis_real_loss = self.adversarial_loss(dis_real, True, True)
        dis_fake_loss = self.adversarial_loss(dis_fake, False, True)
        dis_loss += (dis_real_loss + dis_fake_loss) / 2


        # unet loss
        unet_input_fake = torch.cat((rgb_images.cuda(), outputs_dil.cuda()), dim=1).cuda()
        unet_fake, unet_fake_feat = self.discriminator_dil(unet_input_fake)
        unet_gan_loss = self.adversarial_loss(unet_fake, True, False) * 1.0
        unet_loss_dil += unet_gan_loss

        unet_criterion = self.criterion(outputs_dil.cuda(), mask_images.cuda())
        unet_loss_dil += unet_criterion

        return unet_loss_dil, dis_loss_dil

    def backward(self, unet_loss, dis_loss):
        dis_loss.backward()
        self.dis_optimizer.step()

        unet_loss.backward()
        self.unet_optimizer.step()

    def backward_dil(self, unet_loss_dil, dis_loss_dil):
        dis_loss_dil.backward()
        self.dis_dil_optimizer.step()

        unet_loss_dil.backward()
        self.unet_dil_optimizer.step()

    def predict(self, rgb_images, edge_images):
      input_unet = torch.cat((rgb_images.cuda(), edge_images.cuda()), dim=1).cuda()
      predict_mask = self.unet(input_unet)

      input_unet_dil = torch.cat((rgb_images.cuda(), predict_mask.cuda()), dim=1).cuda()
      predict_mask_dil = self.unet_dil(input_unet_dil)
      return predict_mask_dil

    def predict(self, rgb_images):
      predict_mask = self.unet(rgb_images.cuda())
      return predict_mask

    def train(self, max_epoch, max_epoch_dil):
      i = 0
      for epoch in range(max_epochs):
          for rgb_images, mask_images, edge_images in train_loader:
              unet_loss, dis_loss = self.process(rgb_images.cuda(), mask_images.cuda(), edge_images.cuda())
              self.backward(unet_loss.cuda(), dis_loss.cuda())
              print(str(i) + " " + str(unet_loss.cpu().item())+ " : " + str(dis_loss.cpu().item()))
              i += 1

      i = 0
      for epoch in range(max_epochs):
          for rgb_images, mask_images, edge_images in train_loader:
              input_unet_dil = torch.cat((rgb_images.cuda(), edge_images.cuda()), dim=1).cuda()
              outputs_dil = self.unet(input_unet_dil)
              unet_loss_dil, dis_loss_dil = self.process_dil(rgb_images.cuda(), mask_images.cuda(), edge_images.cuda())
              self.backward_dil(unet_loss_dil.cuda(), dis_loss_dil.cuda())
              print(str(i) + " " + str(unet_loss.cpu().item())+ " : " + str(dis_loss.cpu().item()))
              i += 1
```
